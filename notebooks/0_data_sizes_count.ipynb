# Maximum sizes of corpora (10%) from cc100
As shown at https://data.statmt.org/cc-100/ 6 languages

```python
C = {'en': 8.20e9, 
     'es': 1.40e9,
     'el': 0.74e9,
     'zh': 1.4e9,
     'tr': 0.54e9,
     'ar': 0.54e9,
     'sw': 0.0332e9
    }
```

## Compute the sizes for balancing coeficient
alpha 0, 0.25, 0.5, 0.75, 1.0

with the following eauation:

$c_L = c_{min} \cdot (\frac{|C_L|}{c_{min}})^\alpha$

```python
c_min = 30e6

def compute_corpus_size(alpha, C_L):
    return float('{:0.2e}'.format(c_min * pow((C_L / c_min),  alpha)))
```

```python
corpus_sizes = {}
alphas = [0.0, 0.25, 0.5, 0.75, 1.0]

for L, C_L in C.items():
    corpus_sizes[L] = {}
    for alpha in alphas:
        corpus_sizes[L][f"alpha{alpha}"] = compute_corpus_size(alpha, C_L)
```

```python
corpus_sizes
```

## Plot corpora

```python
plt.figure()

for lang, sizes in corpus_sizes.items():
    plt.plot(alphas, sizes.values(), label=lang, alpha=0.6)
    
#plt.yscale("log")   
plt.legend()
```
